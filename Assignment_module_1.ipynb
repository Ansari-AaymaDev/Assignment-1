 5. Your younger brother opens his phone browser and types "www.flipkart.com".
Within 2 seconds, the Flipkart homepage appears on his screen. He asks you — "How did this
happen so fast? Where did this page come from?"
Explain the complete journey — from the moment he types the website name to the moment the
page shows up on his screen. Use simple terms like DNS, server, browser, HTTP/HTTPS in
your explanation.
- I would tell my brother that when he types Flipkart’s website address in the browser, a lot of things happen very quickly behind the scenes.

- First, the browser takes the website name (www.flipkart.com
) and sends a request to something called Domain Name System. DNS works like a phonebook of the internet. It converts the domain name into an IP address, because computers and servers understand only numbers, not names.

- Once the IP address is found, the browser sends a request over the internet using HTTP or HTTPS (which is the secure version).

- This request reaches the server, which is a powerful computer where Flipkart’s website data is stored (like images, text, and code).

- The server then sends back the required files (HTML, CSS, JavaScript) to the browser.

- Finally, the browser reads these files and renders the webpage on the screen, so we can see and interact with it.

 7. What is the difference between a compiler and an interpreter? Explain in your own
words with a simple real-life comparison.
- A compiler and an interpreter are both used to convert the code we write into a form that the computer can understand, but they do it in different ways.

- A compiler translates the entire program at once before running it. If there are errors, it shows them after checking the whole code. Once the program is compiled successfully, it runs faster because everything is already translated.

- An interpreter, on the other hand, translates and runs the code line by line. If it finds an error, it stops immediately at that line and shows the error. This makes it easier to find mistakes, but execution is usually slower.
- example: A compiler is like translating the whole book first, and then reading it.
- An interpreter is like translating each sentence one by one while reading.


 8. What is the Software Development Life Cycle (SDLC)? Name all its phases and
write one line about what happens in each phase.
- The Software Development Life Cycle (SDLC) is a step-by-step process followed to develop software in a planned and organized way, from idea to final product.

Phases of SDLC:
- Planning – Decide what the project is about, its goals, budget, and timeline.
- Requirement Analysis – Understand and collect what the user or client actually needs.
- Design – Create the structure of the software, like how it will look and how different parts will work.
- Development (Coding) – Write the actual code to build the software.
- Testing – Check for bugs and errors to make sure everything works correctly.
- Deployment – Release the software so users can start using it.
- Maintenance – Fix issues, update features, and improve the software over time.


 9. What is the difference between frontend development and backend development?

- The difference between frontend development and backend development is mainly about what users can see and what happens behind the scenes.

- Frontend development is the part of the application that users interact with directly. It includes things like buttons, layouts, images, and pages. It focuses on how the app looks and feels.

- Backend development is the part that works behind the scenes. It handles things like storing data, processing requests, managing users, and connecting to the database. Users cannot see this part, but it is very important for the app to work properly.
 Let’s take an example of Flipkart:

- When we open the app and see products, search for items, and click on “Buy Now”, that is the frontend.

- But when we place an order:

- The app sends our request to the backend
- The backend checks product availability
- It saves our order details in the database
- It processes payment and sends confirmation

- Then the backend sends the result back to the frontend, and we see “Order Placed Successfully” on our screen.

 10. What is the difference between HTTP and HTTPS? Why is HTTPS important
when you are using a banking or payment website?
- The difference between HTTP and HTTPS is mainly about security.

- HTTP (HyperText Transfer Protocol) is used to transfer data between the browser and the server, but the data is sent in plain text. This means if someone intercepts the data, they can read it easily.

- HTTPS (HyperText Transfer Protocol Secure) is the secure version of HTTP. It uses encryption (through SSL/TLS) to protect the data being sent. This means even if someone tries to intercept the data, they cannot understand it.

- Why HTTPS is important for banking or payment websites:

- When you use banking or payment websites, you enter sensitive information like:

  - Passwords
  - Card details
  - OTPs

- If the website uses HTTP, this information can be stolen by hackers. But with HTTPS:

- Data is encrypted
- Your information stays private and secure
- It protects against hacking and data theft
